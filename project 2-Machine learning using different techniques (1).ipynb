# Train a linear model for price prediction

```python
from sklearn.linear_model import LinearRegression
linear_model =LinearRegression(normalize=True,fit_intercept=False).fit(x_train, y_train)
```

```python
print("Training_score:" , linear_model.score(x_train,y_train))  #Training score should be above 60%.keep running till it improves or else take another look at your selections
```

Testing score should be > training score.If training score >testing score  it is overfitting.underfitting:Testing score >training score(extremely low)

```python
predictors=x_train.columns
predictors
```

```python
coef=pd.Series(linear_model.coef_,predictors).sort_values()
print(coef)
#coefficient and Intercept value that built the linear model
```

```python
y_pred=linear_model.predict(x_test) #using xtest to predict y values after training
```

```python
df_pred_actual=pd.DataFrame({'predicted': y_pred,'actual':y_test})
```

```python
df_pred_actual.head(10)
```

```python
from sklearn.metrics import r2_score
print("Testing_score : " ,r2_score(y_test,y_pred))
```

# LOGISTIC REGRESSION

```python
median =housing_data['median_house_value'].median()
median
```

```python
housing_data['above_median'] = (housing_data['median_house_value'] - median) > 0 #gets data above the median range and makes a new column
```

```python
housing_data.sample(5)
```

```python
x=housing_data.drop(['median_house_value', 'above_median'],axis=1)
y=housing_data['above_median']
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size=0.2)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

# TRain logistic regression model for price classification

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(solver='liblinear').fit(x_train,y_train)
```

```python
print("Training_score :" ,logistic_model.score(x_train,y_train))
```

```python
y_pred =logistic_model.predict(x_test)
```

```python
df_pred_actual =pd.DataFrame({'predicted':y_pred, 'actual': y_test})
df_pred_actual.head(10)
```

```python
from sklearn.metrics import accuracy_score
print("testing_score :", accuracy_score(y_test,y_pred))
```

# using unsupervised Learning:K-Means Algorithm

```python
from sklearn.cluster import KMeans

#initailize model with 2 parameters ...no of clusters and random
Kmeans_model=KMeans(n_clusters=5, random_state=1)
#Get only the numeric columns from housing_data
house =housing_data._get_numeric_data()
#Fit the Model using the house columns
Kmeans_model.fit(house)
#Get the cluster assignment
labels=Kmeans_model.labels_
```

we'll have to reduce the dimensionality of our data without losing too much information.On way to do this is a technique called principal component analysis or PCA.PCA takes multiple columns, and turns them into fewer columns while trying to preserve the unique information in each column

```python
#import the PCA model
from sklearn.decomposition import PCA
#create PCA model
pca_2=PCA(2)
#Fit the PCA model on the numeric columns from earlier
plot_columns =pca_2.fit_transform(house)
#Make a scatter plot of each house ,shaded accordingly to cluster assignment
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=labels)
#show the plot
plt.show()
```

the graph has clustered the houses into different price ranges
